```python
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv'
```

```python
dado = np.loadtxt(url, delimiter=',', usecols=np.arange(1, 6, 1), skiprows=1)
dado
```

```python
diameterOrange = dado[0:5000, 0]
```

```python
weightOrange = dado[:5000, 1]
```

```python
diameterGrapefruit = dado[5001:, 0]
```

```python
weightGrapefruit = dado[5001:, 1]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(diameterOrange, weightOrange)
plt.plot(diameterGrapefruit, weightGrapefruit)
```

```python
xOrange = diameterOrange
yOrange = weightOrange
xGrapefruit = diameterGrapefruit
yGrapefruit = weightGrapefruit
nO = np.size(xOrange)
nG = np.size(xGrapefruit)
```

```python
aO = (nO*np.sum(xOrange*yOrange)-np.sum(xOrange)*np.sum(yOrange))/(nO*np.sum(xOrange**2)-np.sum(xOrange)**2)
bO = np.mean(yOrange) - aO*np.mean(xOrange)
aG = (nG*np.sum(xGrapefruit*yGrapefruit)-np.sum(xGrapefruit)*np.sum(yGrapefruit))/(nG*np.sum(xGrapefruit**2)-np.sum(xGrapefruit)**2)
bG = np.mean(yGrapefruit) - aG*np.mean(xGrapefruit)
```

```python
yO = aO*xOrange+bO
yG = aG*xGrapefruit+bG
```

```python
np.linalg.norm(xOrange-yO)
```

```python
np.linalg.norm(xGrapefruit-yG)
```

```python
plt.plot(xOrange, yOrange)
plt.plot(xOrange, yO)
```

```python
plt.plot(xGrapefruit, yGrapefruit)
plt.plot(xGrapefruit, yG)
```
