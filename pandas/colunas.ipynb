```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
```

```python
dado = pd.read_csv(url, sep=';')
```

```python
dado['Valor/mes'] = dado['Valor'] + dado['Condominio']
```

```python
dado['Valor/ano'] = dado['Valor/mes']*12+dado['IPTU']
```

```python
dado['Descricao'] = dado['Tipo']+' em '+dado['Bairro']+' com '+dado['Quartos'].astype(str)+\
                    ' quarto(s)'+' e '+dado['Vagas'].astype(str)+' vagas de garagem '
```

```python
dado['PossuiSuite'] = dado['Suites'].apply(lambda x: "sim" if x > 0 else "n√£o")
```

```python
dado.to_csv('dado_completo_dev.csv', index=False, sep=';')
```
