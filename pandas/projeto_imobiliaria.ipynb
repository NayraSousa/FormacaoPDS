```python
dados_rejeitados = ['Conjunto Comercial/Sala', 'Prédio Inteiro', 'Loja/Salão', 
                    'Galpão/Depósito/Armazém', 'Casa Comercial', 'Terreno Padrão', 
                    'Box/Garagem', 'Loja Shopping/ Ct Comercial', 'Loteamento/Condomínio', 
                    'Hotel', 'Indústria', 'Sítio', 'Pousada/Chalé', 'Chácara']
```

query() -> seleciona linhas específicas a partir de uma condição específica

```python
casas = dados.query('@dados_rejeitados not in Tipo')
```

```python
casas.head()
```

```python
casas.Tipo.unique()
```

```python
preco_casas = casas.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
preco_casas.plot(kind='barh', figsize=(14, 10), color='green')
```

value_count() -> vai contar a quantidade de cada tipo

```python
casas.Tipo.value_counts()
```

adicionando o normaliza=True, ele vai mostrar o percentual e o to_frame vai transformar em um dataframe

```python
casas.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')
```

```python
perc_casas = casas["Tipo"].value_counts(normalize=True).to_frame().sort_values('Tipo')
perc_casas.plot(kind='bar', figsize=(14, 10), color='green', 
                edgecolor='black', xlabel='Tipo', ylabel='Percentual')
```
